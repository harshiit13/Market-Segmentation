```python
from scipy.spatial.distance import pdist
from scipy.cluster.hierarchy import linkage, dendrogram





risk_dist = pdist(data, metric='cityblock')


risk_hcl = linkage(risk_dist, method='complete')

plt.figure(figsize=(10, 7))
dendrogram(risk_hcl)
plt.title('Hierarchical Clustering Dendrogram (Complete Linkage)')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```

```python

from scipy.cluster.hierarchy import fcluster, linkage, dendrogram




risk_dist = pdist(data, metric='cityblock')

risk_hcl = linkage(risk_dist, method='complete')


c2 = fcluster(risk_hcl, t=20, criterion='distance')

cluster_counts = np.bincount(c2)

print("Cluster Counts:")
print(cluster_counts)
```

```python

from scipy.cluster.hierarchy import fcluster, linkage, dendrogram
import matplotlib.pyplot as plt


risk_dist = pdist(data, metric='cityblock')

risk_hcl = linkage(risk_dist, method='complete')


c6 = fcluster(risk_hcl, t=20, criterion='distance')


data['Cluster'] = c6


c6_means = data.groupby('Cluster').mean()

c6_means_rounded = c6_means.round(1)

print("Cluster Means (rounded to 1 decimal place):")
print(c6_means_rounded)
```

```python
# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(c6_means_rounded.index, c6_means_rounded['Age'], color='skyblue')
plt.xlabel('Cluster')
plt.ylabel('Mean Age')
plt.title('Mean Age by Cluster')
plt.xticks(np.arange(1, len(c6_means_rounded.index) + 1))
plt.grid(axis='y')
plt.show()
```
